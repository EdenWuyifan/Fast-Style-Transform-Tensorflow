```python
import argparse 
import tensorflow as tf
import utils
import numpy as np
import cv2

def load_graph(frozen_graph_filename):
    # parsing the graph_def file
    with tf.gfile.GFile(frozen_graph_filename, "rb") as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
    
 
    # loading the graph_def in the default graph
    with tf.Graph().as_default() as graph:
        tf.import_graph_def(
            graph_def, 
            input_map=None, 
            return_elements=None, 
            name="prefix", 
            op_dict=None, 
            producer_op_list=None
        )
    return graph
 
if __name__ == '__main__':
    frozen_model = "Trained Model/frozen_model.pb"
    graph = load_graph(frozen_model)
 
    # list all operations
    #for op in graph.get_operations():
        #print(op.name,op.values())
    #为了预测,我们需要找到我们需要feed的tensor,那么就需要该tensor的名字
    #注意prefix/Placeholder/inputs_placeholder仅仅是操作的名字,prefix/Placeholder/inputs_placeholder:0才是tensor的名字
    x = graph.get_tensor_by_name('prefix/input:0')
    y = graph.get_tensor_by_name('prefix/clip_by_value:0')
    
    img_path = "content/Starrynight.jpg"
    input_img = utils.load_image(img_path)
    
    with tf.Session(graph=graph) as sess:
        y_out = sess.run(y, feed_dict={x:input_img})
        
    image = cv2.cvtColor(np.array(y_out), cv2.COLOR_RGB2BGR)
    cv2.imwrite("outputs/output.jpg", image)
    print("Output image is saved to 'outputs/output.jpg'")
        
    #utils.save_image(y_out, "test.jpg")
```
